```python
class Human:
    last_name = ''

    def introduce(self):
        print(f"Hello! My name is {self.last_name}")


class Developer(Human):
    language = ""
    company = ''
    IDE = ''

    def make_some_code(self):
        self.introduce()
        return f"I am a {self.language} in {self.company} company."


class PythonDeveloper(Developer):
    tech = ['OOP', 'Flask', 'Django']


Kovalyuk = PythonDeveloper()
Kovalyuk.company = 'GoIt'
Kovalyuk.IDE = 'VS Code'
Kovalyuk.language = 'Python'
Kovalyuk.last_name = 'Dmytro Kovalyuk'

Kovalyuk.make_some_code()
```

```python
class Bird:
    def __init__(self):
        print("Bird is ready")

    def whoisThis(self):
        print("Bird")

    def swim(self):
        print("Swim faster")


class Penguin(Bird):
    def __init__(self):
        # call super() function
        super().__init__()
        print("Penguin is ready")

    def whoisThis(self):
        print("Penguin")

    def run(self):
        print("Run faster")


peggy = Penguin()
peggy.whoisThis()
peggy.swim()
peggy.run()
```

```python
class Human:
    first_name = ''
    last_name = ''
    gender = None
    age = 0


class Student(Human):
    city = ''
    rating = 0

    def set_rating(self, new_rating):
        self.rating = new_rating


class Group:
    title = ''  #UA_Python_1y_10
    cource = '' #Python
    students = []
    

    def print_students(self):
        for curent_student in self.students:
            print('Name: ', curent_student.last_name, curent_student.rating)

    def find_student(self, find_name):
        for curent_student in self.students:
            if curent_student.last_name == find_name:
                print('Name: ', curent_student.last_name, curent_student.rating)
                break
        else:
            print('Not found')

    def add_student(self, new_student):
        self.students.append(new_student)


best_group = Group()
best_group.title = 'UA_Python_1y_10'
best_group.cource = 'Python'

Kovalyuk = Student()
Kovalyuk.city = 'Kyiv'
Kovalyuk.last_name = 'Kovalyuk Dmytro'
Kovalyuk.set_rating(99)

Kovalyuk_I = Student()
Kovalyuk_I.city = 'Kyiv'
Kovalyuk_I.last_name = 'Kovalyuk Ihor'
Kovalyuk_I.set_rating(96)

best_group.add_student(Kovalyuk)
best_group.add_student(Kovalyuk_I)

best_group.print_students()
```

```python
class Dog:
    breed = ''      #порода
    nick = ''       #кличка
    age = 0
    weight = 0      #вага


class Ground:
    dogs = []

    def set_dogs(self, dogs):
        self.dogs = dogs

    def get_dog_with_max_weight(self):
        dog_with_max_weight = self.dogs[0]
        for current_dog in self.dogs:
            if current_dog.weight > dog_with_max_weight.weight:
                dog_with_max_weight = current_dog

        return dog_with_max_weight

     

dog_1 = Dog()
dog_1.breed = 'buldog'
dog_1.nick = 'Barbos'
dog_1.age = 1
dog_1.weight = 16

dog_2 = Dog()
dog_2.breed = 'labrador'
dog_2.nick = 'Orange'
dog_2.age = 2
dog_2.weight = 25

dog_3 = Dog()
dog_3.breed = 'taxa'
dog_3.nick = 'Jack'
dog_3.age = 1
dog_3.weight = 10

dogs = [dog_1, dog_2, dog_3]

volyer = Ground()
volyer.set_dogs(dogs)

dog = volyer.get_dog_with_max_weight()
print(dog.nick, dog.weight)
```
